```python
df=pd.read_csv('movies_updated.csv',index_col="name" )
df
```

```python
print( df.loc['Dr. Dolittle 2']) # displayed by the name indexing
```

```python
print( df.loc['Dr. Dolittle 2' ,["year","budget"]]) # displayed by the name indexing and selecting specific columns to display
```

```python
print( df.loc['Birthday Girl':'Dr. Dolittle 2',["year","budget"]]) # displayed by the name indexing
```

```python
print( df.iloc[0:11 , 0:3]) # displayed by the indexing like an array
```

```python
#SELECTION using pandas
movie =input ('enter the movie name')
try :
    print(df.loc[movie])
except KeyError:
    print(f'{movie} not found')
```

```python
# Filtering by condition using pandas
recentMovies = df[df["year"]>2000]
print(recentMovies['genre'])
```

```python
#aggregation
print("Mean:")
print(df.mean(numeric_only = True)) # we can set all numeric value columns
print("\nsum score:")
print(df['score'].sum()) # we can set specific column
print("\nminimum score:")
print(df['score'].min())
print("\nmaximum vote:")
print(df['votes'].max())
```
