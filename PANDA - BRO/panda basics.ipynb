```python
# Series = one dimensional labeled array/single column that can contain any datatype
```

```python
# example 1
data=[100,101,102,103]
series = pd.Series(data)
series
```

```python
print(series.loc[1])
```

```python
print(series[series>=102])
```

```python
#example 2 using dictionaries
calories= {'day1':1750 ,'day2':2100 ,'day3':1700 }
series =pd.Series(calories)
series
```

```python
# excersise 1
data=['bulbasaur','lvysaur','venusaur','charmander','charmeleon','charizard']
series = pd.Series(data,index=[1,2,3,4,5,6])
series
```

```python
# dataframe : 2 dimentional tabular data structure with rows and columns 
```

```python
data = {
    'name':['abebe','kebede','chala'],
    'age' : [30,35,59]
       }
df =pd.DataFrame(data)
df
```

```python
df =pd.DataFrame(data , index=[101,102,103])
df
```

```python
# accesing a row using assigned index
print(df.loc[102])
```

```python
# accesing a row using index number as an array
print(df.iloc[1])
```

```python
# adding a new column
df['job'] =['artist','coder','N/A']
df
```

```python
# adding a new ROW
newRow =pd.DataFrame([{ 'name':'hiwot' , 'age':'28' , 'job':'musician'}] , 
                    index=[104]
                    )
df=pd.concat([df,newRow])
df
```

```python
newRows =pd.DataFrame([{ 'name':'kal' , 'age':'34' , 'job':'engineer'},
                      { 'name':'john' , 'age':'38' , 'job':'developer'}
                      ] , 
                    index=[105,106]
                    )
df=pd.concat([df,newRows])
df
```
