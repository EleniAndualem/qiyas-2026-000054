```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.set_theme()
tips = sns.load_dataset("tips") # Load an example dataset
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

```python
df = pd.DataFrame({
    "hours": [1, 2, 3, 4, 5, 6, 7],
    "score": [45, 55, 65, 72, 80, 88, 95]
})

graph= sns.relplot(data=df, x="hours", y="score" )
graph.fig.suptitle('Study Hours vs Exam Score')
```

```python
df = pd.DataFrame({
"hours": [1,2,3,4,5,6,7,8],
"score": [50,60,68,75,80,85,92,98],
"gender": ["Male","Male","Male","Male","Female","Female","Female","Female"]
})

graph = sns.relplot(data=df, x="hours", y="score" , hue='gender' , style = 'gender')
graph.fig.suptitle('Exam Scores by Gender')
```

```python
df = pd.DataFrame({
"hours": [1,2,3,4,5,6],
"score": [48,58,67,74,83,90],
"assignments": [2,4,3,6,5,8]
})
graph = sns.relplot(data=df, x="hours", y="score" , size ='assignments')
graph.fig.suptitle('Effect of assignments on scores')
```

```python
df = pd.DataFrame({
"hours": [1,2,3,4,1,2,3,4],
"score": [50,60,70,80,55,65,75,85],
"department": ["CS","CS","CS","CS","IS","IS","IS","IS"]
})
graph = sns.relplot(data=df, x="hours", y="score" ,col='department')
graph.fig.suptitle('Department comparision')
```

```python
df = pd.DataFrame({
"experience": [1,2,3,4,5,6,7,8],
"salary": [30000,35000,42000,50000,58000,65000,72000,80000]
})
graph= sns.lmplot(data=df , x='experience',y='salary')
graph.fig.suptitle('Experince vs Salary')
```
