```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Exercise 1: Monthly Sales Trend
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [120, 150, 170, 160, 190]

plt.plot(months, sales, marker='s', color='green', ls='--', ms=8, label='Sales')
plt.title ('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
#Exercise 2: Student Scores
subjects = ['Math', 'Physics', 'Chemistry', 'Biology']
scores = [85, 78, 92, 88]

plt.plot(months, sales, marker='*', mfc='yellow' , color='red', ls='--', lw = 3 , label='Scores')
plt.title ('Exam  scores')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.legend()
plt.show()
```

```python
#Exercise 3: Company Revenue
years = [2021, 2022, 2023, 2024]
revenue = [500, 700, 850, 1000]

plt.plot(years, revenue , marker='d' , color='blue', ls=':', label='revenue')
plt.title ('Company Revenue')
plt.xlabel('years')
plt.ylabel('revenue')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt  # Import matplotlib library

#Exercise 4: Fruit Supply (Vertical Bar Chart)
fig, ax = plt.subplots()
fruits = ['Apple', 'Orange', 'Banana', 'Mango']
counts = [50, 70, 40, 60]

colors = ['red', 'orange', 'yellow', 'green']
barColors=['tab:red' ,'tab:orange' ,'tab:yellow' ,'tab:green']

ax.bar(fruits , counts, color = barColors , label= "Quantity")
ax.set_ylabel('Quantity')
ax.set_xlabel('fruits')
ax.set_title('Fruit Supply')
ax.legend()
plt.show()
```

```python

sales=pd.DataFrame({
    'day' : ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'cups_sold' : [120,135,110,145,160,210,190]
})

fig, ax = plt.subplots()
ax.bar(sales['day'], sales['cups_sold'],color='green',label='cups_sold')
ax.set_ylabel('cups_sold')
ax.set_title('Fruit by kind and color')
ax.legend()
plt.show()
```
