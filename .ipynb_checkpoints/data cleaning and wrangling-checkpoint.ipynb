```python
import pandas as pd
import numpy as np
```

```python
# Removing Duplicate Records : Tasks
#1. Identify duplicate rows.
#2. Count duplicate rows.
#3. Remove duplicates.
#4. Keep only the last occurrence.
df = pd.DataFrame({
"student_id": [101, 102, 103, 102, 104, 101],
"name": ["Abel", "Sara", "John", "Sara", "Helen", "Abel"],
"score": [80, 90, 75, 90, 88, 80]
})
```

```python
print(df.duplicated())
print("\nduplicate count :", df.duplicated().sum())
```

```python
df_clean = df.drop_duplicates()
df_last = df.drop_duplicates(keep='last')
print (df_clean)
print (df_last)
```

```python
#Duplicate Emails Tasks.
df = pd.DataFrame({
"id":[1,2,3,4],
"email":[
"a@gmail.com",
"b@gmail.com",
"a@gmail.com",
"c@gmail.com"
]
})

#1. Find duplicate emails
duplicates = df[df.duplicated(subset=["email"])]

print(duplicates)
```

```python
#2. Remove duplicate emails.
clean_df = df.drop_duplicates(subset=["email"])
print(clean_df)
```

```python
# Standardize City Names
df = pd.DataFrame({
"city":[
"addis ababa",
"ADDIS ABABA",
"Addis Ababa",
"adDis abAba"
]
})
```

```python
#1. Convert to uppercase.
df["upper"] = df["city"].str.upper()

#2. Convert to lowercase.
df["lower"] = df["city"].str.lower()

#3. Convert to title case.
df["title"] = df["city"].str.title()

print(df)
```

```python
# Remove Extra Spaces
df = pd.DataFrame({
"name":[
" Abel",
"Sara ",
" John ",
" Helen"
]
})

df["name"] = df["name"].str.strip()

print(df)
```

```python
# Standardize Dates
df = pd.DataFrame({
    "date":[
        "2026-01-05",
        "05/02/2026",
        "March 10, 2026",
        "2026.04.15"
    ]
})

#Tasks
#1. Convert to datetime.
#2. Standardize to YYYY-MM-DD.
#3. Extract year.
#4. Extract month.
    
df["date"] = pd.to_datetime(df["date"], format="mixed", dayfirst=False, errors="coerce")
df["formatted"] = df["date"].dt.strftime("%Y-%m-%d")
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month

print(df)
```

```python
#Kilograms to Pounds
df = pd.DataFrame({
"weight_kg":[55,70,85,100]
})

df["weight_lb"] = (df["weight_kg"] * 2.20462).round(2)
print(df)
```

```python
#Celsius to Fahrenheit
df = pd.DataFrame({
"temp_c":[0,25,37,100]
})
df["temp_f"] = (df["temp_c"] * 9/5) + 32

print(df)
```

```python
# Handling Missing Values
df = pd.DataFrame({
"name":["Abel","Sara",None,"John"],
"age":[20,np.nan,25,30]
})
```

```python
#Find missing values.
df.isnull()
```

```python
df.isnull().sum()
```

```python
df["age"] = df["age"].fillna(df["age"].mean())
df["name"] = df["name"].fillna("Unknown")
df
```

```python
#Correct Data Entry Errors
df = pd.DataFrame({
"gender":[
"Male",
"male",
"MALE",
"Female",
"female"
]
})

#Standardize gender values.
df["gender"] = df["gender"].str.title()

print(df)
```

```python
#Currency Conversion
df = pd.DataFrame({
"salary_usd":[500,800,1200,1500]
})

#Convert USD to ETB Assume 1 USD = 135 ETB.
exchange_rate = 135

df["salary_etb"] = df["salary_usd"] * exchange_rate

print(df)
```

```python
#Outlier Detection
df = pd.DataFrame({
"salary":[3500,4000,4200,3900,4100,45000]
})

Q1 = df["salary"].quantile(0.25)

Q3 = df["salary"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[
(df["salary"] < lower) |
(df["salary"] > upper)
]

print(outliers)

clean_df = df[
(df["salary"] >= lower) &
(df["salary"] <= upper)
]

print(clean_df)
```

```python
# Phone Number Standardization

df = pd.DataFrame({
"phone":[
"0911223344",
"+251911223344",
"251911223344",
"0911-223-344"
]
})

#Remove symbols and Convert all numbers to +251 format.

df["phone"] = df["phone"].str.replace("-", "", regex=False)

df["phone"] = df["phone"].str.replace("+", "", regex=False)
df["phone"] = df["phone"].apply(
lambda x:
"+251" + x[1:]
if x.startswith("0")
else "+251" + x[3:]
)

print(df)
```

```python
# Email Cleaning

df = pd.DataFrame({
"email":[
" ABEL@GMAIL.COM ",
"sara@gmail.com",
"JOHN@YAHOO.COM"
]
})

#1. Remove spaces.
#2. Convert to lowercase.
#3. Validate emails.

df["email"] = df["email"].str.strip()

df["email"] = df["email"].str.lower()

df["valid"] = df["email"].str.contains("@")

print(df)
```

```python
# Product Name Normalization
df = pd.DataFrame({
"product":[
"Laptop",
"laptop",
"LAPTOP",
"Laptop "
]
})

#unique counts before cleaning.
print("Before:", df["product"].nunique())

df["product"] = (
df["product"]
.str.strip()
.str.lower()
)

#unique counts after cleaning.
print("After:", df["product"].nunique())
```

```python
#Mixed Data Cleaning

df = pd.DataFrame({
"name":[" Abel ","Sara",None,"JOHN"],
"age":[20,np.nan,25,30],
"salary":[4000,4500,5000,45000],
"join_date":[
"2026-01-01",
"01/02/2026",
"March 10, 2026",
"2026.04.15"
]
})
df
```

```python
#Tasks
#1. Clean names.
df['name'] =df['name'].fillna("UNKOWN")
#2. Fill missing values.
df['age'] =df['age'].fillna(df['age'].mean())

#3. Standardize dates.
df["join_date"] = pd.to_datetime(df["join_date"], format="mixed", dayfirst=False, errors="coerce")
df["DATE"] = df["join_date"].dt.strftime("%Y-%m-%d")

print (df)

#4. Detect outliers.
Q1 = df["salary"].quantile(0.25)
Q3 = df["salary"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[
(df["salary"] < lower) |
(df["salary"] > upper)
]
print("Outliers",outliers)
```
