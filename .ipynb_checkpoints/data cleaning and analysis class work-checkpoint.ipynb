```python
import pandas as pd
import numpy as np

university= pd.DataFrame({
    "student_id" :[201,202,203,204,205,206,207,208,209,210],
    "name" : ["abel","sara",None,"john","marta",None,"david","helen","tom",None],
    "department" :["CS","IT","CS",None,"SE","IT","SE",None,"CS","IT"],
    "GPA": [3.5,3.8,np.nan ,2.9,3.7,np.nan , 3.2 ,3.4,3.9,np.nan],
    "scholarship": [5000 , np.nan, 3000 ,2000, np.nan,4000 , 3500,np.nan,5000,np.nan]
})
```

```python
# whole table
university
```

```python
# Data Cleaning Question No.1
print("--- Couting the missing values from each column ---")
print (university.isnull().sum())
print(f"the total number of missing values = {university.isnull().sum().sum()}")
```

```python
# Data Cleaning Question No. 2
print("--- calculating the missing values percentage from each column ---")
(university.isnull().sum()/len(university))*100
```

```python
# Data Cleaning Question No. 3
print("--- replace missing names by unknown  ---")
university['name']= university['name'].fillna("unknown")
print("--- university table after we replaced missing names by unknown ---")
university
```

```python
# Data Cleaning Question No. 4
print("--- replace missing departments using mode  ---")
university['department']= university['department'].fillna(
    university['department'].mode()[0]
)
print("--- university table after we replaced missing departments by the common department ---")
university
```

```python
# Data Cleaning Question No. 5
print("--- replace missing gpa using median  ---")
university["GPA"] = university["GPA"].fillna(
university["GPA"].median()
)
print("--- university table after we replaced missing GPA by the Median of ther students GPA ---")
university
```

```python
# Data Cleaning Question No. 6
print("--- replace missing SCHOLARSHIPS with ZERO  ---")
university["scholarship"] = university["scholarship"].fillna(0)
print("--- university table after we replacedSCHOLARSHIPS with ZERO  ---")
university
```

```python
# Data Analysis Question No. 1
print("--- counting students by department  ---")
counts={}
for i , row in university.iterrows():
    deppartmentName = row['department']
    counts[deppartmentName] = counts.get(deppartmentName,0)+1
print(counts)
```

```python
# Data Analysis Question No. 2
print("--- counting scholarship recipients ---")
for i, row in university.iterrows():
    if row['scholarship'] > 0.0:
        print(row["name"])
```

```python
# Data Analysis Question No. 3
print("--- Finding the highest GPA ---")
highestGPA = university['GPA'].max()

# Find the student(s) with that GPA
studentsWithHighestGPA = university[university['GPA'] == highestGPA]

for index, row in studentsWithHighestGPA.iterrows():
    print(f"Name - {row['name']} \nGPA = {row['GPA']}")
```

```python
# Data Analysis Question No. 4
print("--- Finding the lowest GPA ---")
lowestGPA = university['GPA'].min()

# Find the student(s) with that GPA
studentsWithLowestGPA = university[university['GPA'] == lowestGPA]

for index, row in studentsWithLowestGPA.iterrows():
    print(f"Name - {row['name']} \nGPA = {row['GPA']}")
```

```python
# Data Analysis Question No. 5
print("--- average GPA ---")
print (f"mean = {university["GPA"].mean()}")
```

```python
# Data Analysis Question No. 6
print("--- Sum of all Scholarships ---")
print (f"Total scholarships = {university["scholarship"].sum()}")
```
