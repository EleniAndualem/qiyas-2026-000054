```python
import requests 
from bs4 import BeautifulSoup
```

```python
htmlDoc= requests.get("https://www.worldometers.info/coronavirus/")
soup = BeautifulSoup(htmlDoc.content , 'html.parser')
```

```python
tableContent= soup.find('table')
```

```python
tableHead =tableContent.find('thead')
tableBody=tableContent.find('tbody')
tableRow=tableBody.find_all('tr')

for data in tableRow[8:]:
    print ( data.text)
    print ("=================================================================")
    
```

```python
file=open("corona data.txt", 'a' ,encoding='utf-8')
```

```python
for data in tableRow[8:]:
    file.write(data.text)
    file.write("=================================================================")
```

```python
file.close()
```
