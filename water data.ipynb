```python
import requests 
from bs4 import BeautifulSoup
```

```python
htmlDoc= requests.get("https://www.worldometers.info/water/")
soup =  BeautifulSoup(htmlDoc.content,'html.parser')
```

```python
tablecontent = soup.find('table')
```

```python
tableHead=tablecontent.find('thead')
tableBody=tablecontent.find('tbody')
tableRow=tableBody.find_all('tr')
```

```python
for header in tableHead:
    print ( " " + header.text + " ")
    print ("=================================================================")
    
```

```python
for data in tableRow:
    print ( data.text)
    print ("=================================================================")
    
```

```python
file=open("water data.txt", 'a' ,encoding='utf-8')
```

```python
for data in tableRow:
    file.write( data.text)
    file.write("=================================================================")
```

```python
for header in tableHead:
    file.write ( " " + header.text + " ")
    file.write ("=================================================================")
    
```

```python
file.close()
```
