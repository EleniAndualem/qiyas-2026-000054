```python
import pandas as pd
import numpy as np

university= pd.DataFrame({
    "student_id" :[201,202,203,204,205,206,207,208,209,210],
    "name" : ["abel","sara",None,"john","marta",None,"david","helen","tom",None],
    "department" :["CS","IT","CS",None,"SE","IT","SE",None,"CS","IT"],
    "GPA": [3.5,3.8,np.nan ,2.9,3.7,np.nan , 3.2 ,3.4,3.9,np.nan],
    "scholarship": [5000 , np.nan, 3000 ,2000, np.nan,4000 , 3500,np.nan,5000,np.nan]
})
```

```python
university[university["GPA"].isnull()]
```

```python
university["name"] = university["name"].fillna("Unknown")
university
```

```python
university["GPA"] = university["GPA"].fillna(
university["GPA"].mean()
)
university
```

```python
university["GPA"] = university["GPA"].fillna(
university["GPA"].median()
)
university
```

```python
university["department"] = university["department"].fillna(
university["department"].mode()[0]
)
university
```

```python
university["scholarship"] = university["scholarship"].fillna(0)
university
```

```python
#Remove rows containing missing values.
university.dropna()
```

```python
#Remove columns containing missing values.
university.dropna(axis=1)
```

```python
#Create a GPA missing-value flag.
university["gpa_missing"] = university["GPA"].isnull()
university
```

```python
#Create a scholarship missing-value flag.
university["scholarship_missing"] = university["scholarship"].isnull()
university
```

```python
#the column with the highest number of missing values.
print(f"max missing = {university.isnull().sum().idxmax()}")
print(f"minimum missing = {university.isnull().sum().idxmin()}")
```
