```python
import time
from bs4 import BeautifulSoup
import pandas as pd
import requests

BBC_URL = "https://www.bbc.com/news/topics/cwlw3xz047jt"
session = requests.Session()
session.headers.update({"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"})
articles = []

for page in range(1, 43):
    print(f"Scraping page {page}...")
    url = BBC_URL if page == 1 else f"{BBC_URL}?page={page}"
    
    try:
        soup = BeautifulSoup(session.get(url).content, "html.parser")
        section = soup.find("section", class_="Section-layouts-styles__SectionLayoutWrapper-sc-40ef2d4a-0 bgdYKK")
        if not section: continue

        for a in section.find_all("a", {"data-testid": "internal-link"}):
            lnk = f"https://www.bbc.com{a.get('href')}" if a.get("href") else None
            img = a.find("img")
            
            # FORCE READ THE REAL IMAGES: Pull from srcset to bypass the JS placeholder trick
            banner = None
            if img:
                srcset = img.get("srcset")
                banner = srcset.split(",")[0].split(" ")[0] if srcset else img.get("src")

            # Article page content extraction
            author, content = "Unknown", "No content found"
            if lnk:
                time.sleep(1.2)
                try:
                    art = BeautifulSoup(session.get(lnk).content, "html.parser").find("article")
                    if art:
                        auth_tag = art.find("div", class_=lambda c: c and "Byline" in c)
                        author = auth_tag.text if auth_tag else "Unknown"
                        content = " ".join([p.text.strip() for p in art.find_all("p")])
                except Exception: 
                    pass

            articles.append({
                "Date": a.find("span").text if a.find("span") else "No Date",
                "Title": a.find("h2").text if a.find("h2") else "No Title",
                "Description": a.find("p").text if a.find("p") else "No Description",
                "Banner": banner, 
                "Article Link": lnk, 
                "Author": author, 
                "Content": content
            })
    except Exception as e:
        print(f"Page {page} failed unexpectedly: {e}")

# Build DataFrame
df = pd.DataFrame(articles)

# Clean up long content for a perfect CSV display look
df_csv = df.copy()
df_csv["Content"] = df_csv["Content"].apply(lambda x: x[:150] + "..." if len(x) > 150 else x)

# Save to your computer
df_csv.to_csv("bbc_news_scraped.csv", index=False, encoding="utf-8-sig")
print(f"Finished! Total articles exported safely: {len(df)}")
```
