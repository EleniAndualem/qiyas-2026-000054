```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
BBC_URL = "https://www.bbc.com/news/topics/cwlw3xz047jt"
```

```python
for page in range(1, 43):
    url = f"{BBC_URL}?page={page}" if page > 1 else BBC_URL
    htmlDoc = requests.get(url)  
    soup = BeautifulSoup(htmlDoc.content, 'html.parser')
    
    allNews = soup.find('section', class_="Section-layouts-styles__SectionLayoutWrapper-sc-40ef2d4a-0 bgdYKK")
    if allNews:
        articles = allNews.find_all('a', {'data-testid': 'internal-link'})
        for article in articles:
            title = article.find('h2').text if article.find('h2') else None
            description = article.find('p').text if article.find('p') else None
            article_link = f"https://www.bbc.com{article.get('href')}" if article.get('href') else None
            Banner = article.find('img').get('src') if article.find('img') else None
            date = article.find('span').text if article.find('span') else None

            print(f"Date: {date}")
            print(f"Title: {title}")
            print(f"Description: {description}")
            print(f"Article Link: {article_link}")
            print(f"Banner image: {Banner}")
            print("-" * 30)
```
