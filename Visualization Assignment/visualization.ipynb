```python
#Task 2: Number of Movies Released Each Year
movieCount = movies.groupby("year")["name"].count().reset_index(name="count")
plt.plot(movieCount["year"], movieCount["count"], marker='d', color='green', ls='--', ms=8, label='Count')

plt.title('Movies Released Each Year')
plt.xlabel('Years')
plt.ylabel('No of movies')
plt.legend()
plt.grid(True) 
```

```python
# ANSWERING THE Questions
# 1. Which year has the most movies?
print("Question 1 \n Year(s) with the maximum number of movies:")
maxMovieYears =movieCount.nlargest(1, "count", keep="all")["year"].to_string(index=False)
print(maxMovieYears)

#2. Does the number of movies increase over time?
print ("Question 2 \n Does the number of movies increase over time?" )
print ("- yes it does" )

#3. What trend do you observe?
print ("Question 3 \n 3. What trend do you observe?" )
print ("- it was increasing untill 1985 then it was constant till 2000 then it decreased on 2001" )
```

```python
# Task 3: Top 10 Movie Genres
topGenres = movies["genre"].value_counts().head(10).reset_index(name="count")

print("Top 10 Genres :")
print(topGenres)
```

```python
#  creating the vertical bar chart.
fig, ax = plt.subplots()
ax.bar(topGenres['genre'], topGenres['count'], color='green', label='movie count')

plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
ax.set_ylabel('COUNT')
ax.set_title('Top 10 Movie Genres')
ax.legend()
plt.show()
```

```python
#Task 4: Top 10 Directors
topDirectors = movies["director"].value_counts().head(10).reset_index(name="count")

print("Top 10 Genres :")
print(topDirectors)
```

```python
#  creating the horizontal bar chart.
fig, ax = plt.subplots()
ax.barh(topDirectors['director'], topDirectors['count'], color='green', label='movie count')

plt.tight_layout()
ax.set_xlabel('COUNT')
ax.set_ylabel('DIRECTORS')
ax.set_title('Top 10 Movie Directors')
ax.legend()
plt.show()
```

```python
# Task 5: Distribution of Movie Ratings
# creating histogram
plt.hist(movies["score"], bins=30, color='green')

plt.title('Distribution of Movie Ratings')
plt.xlabel('Score')
plt.ylabel('Number of Movies')
plt.show()
```

```python
#Task 6: Average Movie Score by Genre
genreGroup = movies.groupby("genre")["score"].mean()
sortedgenres = genreGroup.sort_values(ascending=False)
Table = sortedgenres.reset_index(name="averageScore")
print("Average Score by Genre:")
print(Table)
```

```python
#  creating the bar chart.
fig, ax = plt.subplots()
ax.bar(Table['genre'], Table['averageScore'], color='green', label='avg score')

plt.xticks(rotation=45, ha='right')
plt.tight_layout() 
ax.set_ylabel('Average Score')
ax.set_title('Movie Genres')
ax.legend()
plt.show()
```

```python
Answering the Questions
1. Which genre has the highest average score?
 - History
2. Which genre has the lowest average score?
 - Western
3. Are the most popular genres also the highest rated?
 - no the popular genre comedy is not highest rated 
```

```python
# Task 7: Top 10 Highest Grossing Movies
clean_text = movies["gross"].astype(str).str.replace("$", "", regex=False).str.replace(",", "", regex=False)
movies["gross"] = pd.to_numeric(clean_text, errors="coerce")

highGrossMovies = movies.nlargest(10, "gross")
print("Top 10 Highest Grossing Movies:")
print(highGrossMovies[["name", "gross"]])
```

```python
#  creating the horizontal bar chart.
fig, ax = plt.subplots()
ax.barh(highGrossMovies["name"], highGrossMovies["gross"], color='green')

plt.tight_layout()
ax.set_xlabel('Gross')
ax.set_ylabel('Movies')
ax.set_title('Top 10 Highest Grossing Movies')
plt.show()
```
