```python
import pandas as pd
```

```python
# student table
students = pd.DataFrame({ 
"Student_ID": [1, 2, 3, 4], 
"Student_Name": ["Abel", "Sara", "John", "Helen"] 
})
students
```

```python
# Enrollments table
enrollments = pd.DataFrame({ 
"Student_ID": [1, 2, 5], 
"Course": ["Python", "SQL", "Machine Learning"] 
})
enrollments
```

```python
# inner join
result = pd.merge(
    students ,
    enrollments,
    on="Student_ID",
    how="inner"
)
result
```

```python
# left join
result=pd.merge(
    students,
    enrollments,
    on = "Student_ID",
    how ="left"
)
result
```

```python
# right join
result = pd.merge( 
students, 
enrollments, 
on="Student_ID", 
how="right" 
)
result
```

```python
# outer join 
result = pd.merge( 
students, 
enrollments, 
on="Student_ID", 
how="outer" 
) 
result
```

```python
result["Course"] = result["Course"].fillna("Not Enrolled") 
result["Student_Name"] = result["Student_Name"].fillna("unknown") 
result
```
