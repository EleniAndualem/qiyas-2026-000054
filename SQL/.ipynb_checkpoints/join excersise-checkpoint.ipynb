```python
import pandas as pd
import numpy as np
```

```python
#Dataset 1: Customers
customers = pd.DataFrame({
"Customer_ID": ["C001", "C002", "C003", "C004", "C005", "C006"],
"Customer_Name": ["Abel", "Sara", "John", "Helen", "Samuel",
"David"],
"City": ["Addis Ababa", "Adama", "Jimma", "Bahir Dar", "Hawassa",
"Dire Dawa"]
})
customers
```

```python
#Dataset 2: Orders
orders = pd.DataFrame({
"Order_ID": [101,102,103,104,105,106,107],
"Customer_ID": ["C001","C002","C001","C004","C007","C006","C006"],
"Product":
["Laptop","Phone","Mouse","Keyboard","Tablet","Monitor","Printer"],
"Amount": [1200,800,45,150,650,700,300]
})
orders
```

```python
#Exercise 1 — Customer Purchase Report
purchaseReport= pd.merge(
    customers,
    orders,
    on="Customer_ID",
    how = "inner"
)
purchaseReport
```

```python
Questions
1. How many records are returned?
  - six
2. Which order is excluded from the report?
  - the fifth order
3. Why was that order excluded?
  - because Customer C007 is not registered
```

```python
#Exercise 2 — Order Investigation
purchaseReport= pd.merge(
    customers,
    orders,
    how = "right"
)
purchaseReport
```

```python
Questions
1. Which order has missing customer information?
  - order 105
2. Why is the customer information missing?
  - because Customer C007 is not registered
3. How many registered customers do not appear in this report?
  - two customer C003 and C005 didnt place any order
```

```python
#Exercise 3 — Customer EngagementReport
purchaseReport= pd.merge(
    customers,
    orders,
    how = "left"
)
purchaseReport
```

```python
Questions
1. Which customer has never placed an order?
  - C003 and C005 didnt place any order
2. What values appear for that customer's Product and Amount?
  - NaN
3. Why do those values appear?
  - b/c those customers didnt place any order and instead of empty data NaN values are placed
```

```python
#Exercise 4 — Data Quality Audit
purchaseReport= pd.merge(
    customers,
    orders,
    how = "outer"
)
purchaseReport
```

```python
Questions
1. How many rows are returned?
  - nine
2. Which registered customer has no orders?
  - customer C003 and C005
3. Which order belongs to a customer that is not registered?
   - order 105
4. Why are both records important to keep during a data quality audit?
  - it helps us to compare both customer and order table 
```

```python
# Exercise 5 — Customer Support Investigation
purchaseReport= pd.merge(
    customers,
    orders,
    how = "right"
)
print (purchaseReport[["Customer_Name" , "Product" , "Amount"]])
```
