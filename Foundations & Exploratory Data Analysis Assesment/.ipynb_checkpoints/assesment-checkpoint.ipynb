```python
# Section C: Pandas DataFrames
import pandas as pd
```

```python
# Exercise 14: Create a DataFrame
students = pd.DataFrame({
    "Name" : ["abebe" ,"kebede" ,"chala"],
    "age" : [25,24,22] ,
    "score" : [92,43,84]
})
students
```

```python
#Exercise 15: Selecting Rows and Columns
#Display only the Name column.
print (students.Name)
```

```python
#Display students whose score is greater than 70.
print ("students who have score is greater than 70 are:")
for index, row in students.iterrows():
    if row['score'] > 70:
        print(row['Name'])
```

```python
#Exercise 16: Adding New Columns
students["status"] = np.where(students["score"] >= 50, "pass", "fail")
students
```

```python
#Exercise 17: Summary Statistics
students.describe()
```

```python
#Section D: Data Cleaning
#Exercise 18: Missing Values Detection
students= pd.DataFrame({
    "Name" : [None,"abebe" ,"kebede" ,"chala"],
    "age" : [25,24,np.nan,22] ,
    "score" : [92,np.nan,43,84]
})
#Number of missing values in each column.
print (f'missing values in each column -\n{students.isnull().sum()}')
#Total missing values in the dataset.
print (f'Total missing values -\n{students.isnull().sum().sum()}')
```

```python
#Exercise 19: Handling Missing Values
print ("before \n" ,students)
students["age"] = students["age"].fillna(
    students["age"].mean())
students["scorescore"] = students["score"].fillna(
    students["score"].median())
print ("after \n" ,students) 
```
